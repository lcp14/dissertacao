```python
import pandas as pd
import zipfile
import xml.etree.ElementTree as ET
import ntpath
```

```python
autores = pd.read_csv('../datasets/autores/autores.csv',delimiter="|")
```

```python
codinstituicoes = []
for i in autores[' path ']:
    path = '../'+i.strip()
    uid = ntpath.basename(path).split('.')[0]
    #print (folder,a)
    zip_ref = zipfile.ZipFile(path, 'r')
    zip_ref.extractall()
    zip_ref.close()
    #Lendo XML
    xmlTree = ET.parse('curriculo.xml')
    root = xmlTree.getroot()
    if 'DATA-ATUALIZACAO' in root.attrib.keys():
        data_atualizacao = root.attrib['DATA-ATUALIZACAO']
    for dados in root.iter('RESUMO-CV'):
        resumo = dados.attrib['TEXTO-RESUMO-CV-RH']
    for dados in root.iter('ENDERECO-PROFISSIONAL'):
        codInstituicao = dados.attrib['CODIGO-INSTITUICAO-EMPRESA']
        nomeInstituicao = dados.attrib['NOME-INSTITUICAO-EMPRESA']
    for dados in root.iter('DOUTORADO'):
        docCodInstituicao = dados.attrib['CODIGO-INSTITUICAO']
        docNomeInst = dados.attrib['NOME-INSTITUICAO']
        docAnoInicio = dados.attrib['ANO-DE-INICIO']
        docAnoFim = dados.attrib['ANO-DE-CONCLUSAO']
    for dados in root.iter('POS-DOUTORADO'):
        posCodInstituicao = dados.attrib['CODIGO-INSTITUICAO']
        posNomeInst = dados.attrib['NOME-INSTITUICAO']
        posAnoInicio = dados.attrib['ANO-DE-INICIO']
        posAnoFim = dados.attrib['ANO-DE-CONCLUSAO']
    for dados in root.iter('AREA-DE-ATUACAO'):
        area = dados.attrib['NOME-DA-AREA-DO-CONHECIMENTO']
        subarea = dados.attrib['NOME-DA-SUB-AREA-DO-CONHECIMENTO']
        esp = dados.attrib['NOME-DA-ESPECIALIDADE']
        table_area.append([uid,area,subarea,esp])
        #print(area +'->'+subarea+'->'+esp)
    for id_idioma,dados in enumerate(root.iter('IDIOMA')):
        desIdioma = dados.attrib['DESCRICAO-DO-IDIOMA']
        idioma = dados.attrib['IDIOMA']
        table_idioma.append([uid,id_idioma,uid+str(id_idioma),idioma,desIdioma])
    for dados in root.iter('INFORMACAO-ADICIONAL-INSTITUICAO'):
        #print(dados.attrib['CODIGO-INSTITUICAO'])
        codInstituicao2 = dados.attrib['CODIGO-INSTITUICAO']
        if codInstituicao == codInstituicao2:
            siglaInstituicao = dados.attrib['SIGLA-INSTITUICAO']
            if codInstituicao not in codinstituicoes:
                instituicao.append([codInstituicao,siglaInstituicao,nomeInstituicao])
                codinstituicoes.append(codInstituicao)
        else:
            if docCododInstituicao not in codinstituicoes:
                instituicao.append([codInstituicao,'',docNomeInstituicao])
            if posCodInstituicao not in codinstituicoes:
                instituicao.append([codInstituicao,'',posNomeInstituicao])
    table_autor.append([uid,data_atualizacao,resumo,docCodInstituicao,docAnoInicio,docAnoFim,posCodInstituicao,posAnoInicio,posAnoFim,codInstituicao])
```

```python
pd.DataFrame(table_autor[1:],columns=table_autor[0]).to_csv('autor_info.csv',index=False,sep='|')
pd.DataFrame(table_idioma[1:],columns=table_idioma[0]).to_csv('idioma.csv',index=False,sep='|')
pd.DataFrame(instituicao[1:],columns=instituicao[0]).to_csv('instituicoes.csv',index=False,sep='|')
pd.DataFrame(table_area[1:],columns=table_area[0]).to_csv('autor_area.csv',index=False,sep='|')
```
