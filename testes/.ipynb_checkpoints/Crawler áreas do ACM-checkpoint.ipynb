# Hierarquia de áreas da ACM

Salva um arquivo contendo a hierarquia de áreas da ACM. A partir da página local, já que a ACM bloqueia o acesso para crawler.

A hierarquia é feita com listas dentro de listas.

```python
from bs4 import BeautifulSoup
import requests
import pickle 
path = "acc.html"
f = open(path)


s = BeautifulSoup(f,'html.parser')
tree = {}
```

```python
div_areas = s.find("div", {"id": "holdflat"})
```

```python
uls = div_areas.find_all("ul",recursive=False)
lis = [i for i in div_areas.find_all("'li")]
```

```python
import unicodedata
```

```python
areas = []
def dictify(ul):
    result = {}
    for li in ul.find_all("li", recursive=False):
        key = next(li.stripped_strings).lower()
        ul = li.find("ul")
        print(key)
        if ul:
            result[key] = dictify(ul)
            areas.append(re.sub('\W'," ",unicodedata.normalize('NFKD', key).encode('ascii', 'ignore').decode('utf-8', 'ignore')))
        else:
            result[key] = None
    return result
```

```python
import re
areas_id = {} # key: area value: id
areas_area = {} # key: id value: area
areas_parents = {}
cont = 0
for i in uls[0].find_all('a'):
    if i.has_attr('href'):
        if i.attrs['href'] != '#toc':
            #print(i.attrs['href'].split('("')[1].split('")')[0].split('.')[1:],i.text)
            ids = i.attrs['href'].split('("')[1].split('")')[0].split('.')[1:]
            area = unicodedata.normalize('NFKD', i.text.lower()).encode('ascii', 'ignore').decode('utf-8', 'ignore')
            area = re.sub('\W'," ",area)
            idx = ids[-1]
            
            if idx not in areas_id.keys():
                idx = idx
            else:
                idx = idx+'_'+str(cont)
                cont+=1
            areas_id[idx] = area
            areas_parents[idx] = ids[:-1]
```

```python
def topic_acm_area(topic,areas_id,areas_parents,areas_area):
    m= 0
    similar_areas = []
    for area in areas_id.keys():
        a = fuzz.token_set_ratio(topic,area)
        if a > m:
            m = a

        if a > 70:
            print(area+"-->","-->".join([areas_area[i] for i in areas_parents[areas_id[area]]][::1]))
topic_acm_area(topics[1],areas_id,areas_parents,areas_area)
```

```python
def getKeysByValue(dictOfElements, valueToFind):
    listOfKeys = list()
    listOfItems = dictOfElements.items()
    for item  in listOfItems:
        if item[1] == valueToFind:
            listOfKeys.append(item[0])
    return  listOfKeys
```

```python
getKeysByValue(areas_id,'e commerce infrastructure')
```

```python
def area_parents(area,areas_id,areas_parents):
    li_id = getKeysByValue(areas_id,area)
    parents = [areas_parents[i] for i in li_id]
    for parent in parents:
        print(area+"-->"+"-->".join([areas_id[i] for i in parent][::-1]))
area_parents('e commerce infrastructure',areas_id,areas_parents)
```
