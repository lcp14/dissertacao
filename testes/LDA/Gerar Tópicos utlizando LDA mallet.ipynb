```python
import pandas as pd
import numpy as np
import unicodedata
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.decomposition import NMF

# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

import os

from IPython.display import HTML, display
from tabulate import tabulate
```

```python
mallet_path = '/home/leandro/Mestrado_Organized/mallet-2.0.8/bin/mallet'
```

```python
ano = str(2017)
df_artigos = pd.read_csv('~/Mestrado_Organized/datasets/artigos/preprocessed/2017/artigosPre.csv',sep="|",index_col=0)
#df_artigos = pd.read_csv('datasets/trabalhosPre.txt',sep="|",index_col=0)
artigos = df_artigos['titulo'].values
```

```python
%matplotlib inline
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(artigos)
sum_words = X.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
#print(words_freq)
indices = np.arange(len(words_freq))
word = []
frequency = []
for i in range(len(words_freq)):
    word.append(words_freq[i][0])
    frequency.append(words_freq[i][1])
plt.figure(figsize=(30, 15))
plt.bar(indices,frequency,width =1, color='r')
plt.xticks(indices, word, rotation='vertical')
plt.tight_layout()
plt.show()
```

```python
new_stop_words = []
word = []
frequency = []
indices = 0
for words in words_freq:
    if words[1] <= 10 or words[1] >= 100:
        new_stop_words.append(words[0])
        word.append(words[0])
    elif words[0].isdigit():
        new_stop_words.append(words[0])
        word.append(words[0])
    else:
        frequency.append(words[1])
        indices +=1
plt.figure(figsize=(30, 15))
plt.bar(range(indices),frequency,width =1, color='r')
#plt.xticks(range(indices), word, rotation='vertical')
plt.tight_layout()
plt.show()
new_stop_words.append('datum')
    
```

```python
artigos = [[a for a in artigo.split() if a not in new_stop_words] for artigo in artigos]
```

```python
id2word = corpora.Dictionary(artigos)
corpus = [id2word.doc2bow(artigo) for artigo in artigos]
```

```python
ldamallet = gensim.models.wrappers.LdaMallet(
    mallet_path, corpus=corpus, num_topics=18, id2word=id2word)
```

```python
topics = ldamallet.show_topics(formatted=False)
```

```python
table_output = [[
    'Topico',
    'Palavra 1',
    'Palavra 2',
    'Palavra 3',
    'Palavra 4',
    'Palavra 5',
    'Palavra 6',
    'Palavra 7',
    'Palavra 8',
    'Palavra 9',
    'Palavra 10',
]]
```

```python
for top in topics:
    top_out = [top[0]+1]
    top_out += [i[0] for i in top[1]]
    table_output.append(top_out)
```

```python
display(HTML(tabulate(table_output, tablefmt='html')))
```

```python
# Compute Coherence Score
coherence_model_ldamallet = CoherenceModel(model=ldamallet, texts=artigos, dictionary=id2word, coherence='c_v')
coherence_ldamallet = coherence_model_ldamallet.get_coherence()
print('\nCoherence Score: ', coherence_ldamallet)
```
